Saccades in 4 categories (Wikipedia) :  

* In a **visually guided saccade** , the eyes move toward a visual transient, or stimulus. The parameters of visually guided saccades (amplitude, latency, peak velocity, and duration) are frequently measured as a baseline when measuring other types of saccades. Visually guided saccades can be further subcategorized:   
    + A **reflexive saccade** is triggered exogenously by the appearance of a peripheral stimulus, or by the disappearance of a fixation stimulus.  
    + A **scanning saccade** is triggered endogenously for the purpose of exploring the visual environment.    
* In an **antisaccade**, the eyes move away from the visual onset. They are more delayed than visually guided saccades, and observers often make erroneous saccades in the wrong direction. A successful antisaccade requires inhibiting a reflexive saccade to the onset location, and voluntarily moving the eye in the other direction.  
* In a **memory guided saccade**, the eyes move toward a remembered point, with no visual stimulus.  
* In a **sequence of predictive saccades**, the eyes are kept on an object moving in a temporally and/or spatially predictive manner. In this instance, saccades often coincide (or anticipate) the regularly moving object.

**Saccade Adaptation** - saccades adapt, when brain thinks saccades are too big or small (let by experimental manipulation, or by disorder) the brain corrects.  
[Hopp & Fuchs, 2004. The characteristics and neuronal substrate of saccadic eye movement plasticity] http://www.sciencedirect.com/science/article/pii/S030100820300193X

**Saccadic Masking/Suppression** - It is FALSE that no information is passed from optic nerve to brain during saccade. Low spatial frequencies are attenuated, but high spatial frequencies (fine detail) are unnaffected

**Trans-saccadic prerception** It is also thought that perceptual memory is updated during saccades so that information gathered across fixations can be compared and synthesized. However, the entire visual image is not updated during each saccade, only 3–4 features or objects if they are attended to. Some scientists believe that this is the same as visual working memory, but as in spatial updating the eye movement has to be accounted for. The process of retaining information across a saccade is called trans-saccadic memory, and the process of integrating information from more than one fixation is called trans-saccadic integration.

**[Listings Law] https://en.wikipedia.org/wiki/Listing%27s_law   **
Holds when head is stationary, upright, gaze is directed to far targets. ie fixation, saccades, or smooth pursuit.  
Listing's law states that the eye does not achieve all possible 3D orientations and that, instead, all achieved eye orientations can be reached by starting from one specific "primary" reference orientation and then rotating about an axis that lies within the plane orthogonal to the primary orientation's gaze direction (line of sight / visual axis). This plane is called Listing's plane.

### Engbert & Kliegl, 2004

[Microsaccades Keep the Eyes' Balance During Fixaction](papers/Engbert-Kliegl2004.pdf)

Self-avoiding random walk captures the fundamental properties of these eye movements:  

Eq for generalized version of Einstien's Brownian Motion by MandelBrot & vanNess (1968) aka fractional Brownian motion:  

$$(\Delta x^2) \propto \Delta t^H$$  

also see [Wiki on fractional Brownian Motion](https://en.wikipedia.org/wiki/Fractional_Brownian_motion) ** Something Cool! ** - This relationship is called a 'covariance function', or a 'Kernel'. This is the idea of a mathematical relationship between one element's evolution based on another.

scaling exponent H can be any real number between 0->2. Why? 1 < H < 2 implies trajectory is more likely to go in same direction it went during previous time step (persistance).  0 < H < 1 (antipersistance) means trajectory increments are negatively correlated. H > 2 would mean magnitude is increasing, H < 0 would mean... not really sure. Maybe magnitude is decreasing? In classical BM, H=1 for this equation, & movement at time increments are uncorrelated.  

This paper: Recorded eye movement (500Hz), estimated mean square displacement ($\Delta x^2 $) for each time step using equation: $$ D^2 (m) = \frac{1}{N-m}\sum_{i=1}^{N-m}\|{\vec{x}_{i+m} - \vec{x}_i}\|^2$$  

N = number of samples in recorded time series  
m = time lag (in number of samples)  

1. **Random walk analysis: Drift & Microsaccades. ** Plot the data and solve for slope to get estimate of value for H. in short term 2-20 seconds, H > 1 (persistance), in long term, 20-200 seconds, H < 1 (antipersitance). Compare with randomly shuffled data, where no corrrelation exists, H=1.  

2. **Function of Microsaccades in Binocular Coordination: Remove MS from data. ** Remove microsaccades and reconstruct trajectories based on relative $\Delta t$ for each timepoint. In short time scales, scaling factor decreased from 1.36 to 1.26 on average. In short term, scaling factor for disparity remained close to 1 meaning MS and drift were uncorrelated, but in long term, values of scaling factor for disparity increaased. Microsaccades therefore were responsible for antipersistance in long term.
